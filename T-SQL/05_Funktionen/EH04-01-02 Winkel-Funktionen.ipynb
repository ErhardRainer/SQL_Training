# Winkelfunktionen



<span style="color:rgb(32, 33, 34);font-family:sans-serif;background-color:rgb(255, 255, 255);">Der&nbsp;</span> **Sinus** <span style="color:rgb(32, 33, 34);font-family:sans-serif;background-color:rgb(255, 255, 255);">&nbsp;eines&nbsp;</span> [Winkels](https://de.wikipedia.org/wiki/Winkel) <span style="color:rgb(32, 33, 34);font-family:sans-serif;background-color:rgb(255, 255, 255);">&nbsp;ist das Verhältnis der Länge der&nbsp;</span> [Gegenkathete](https://de.wikipedia.org/wiki/Gegenkathete) <span style="color:rgb(32, 33, 34);font-family:sans-serif;background-color:rgb(255, 255, 255);">&nbsp;(</span>[Kathete](https://de.wikipedia.org/wiki/Kathete)<span style="color:rgb(32, 33, 34);font-family:sans-serif;background-color:rgb(255, 255, 255);">, die dem Winkel gegenüberliegt) zur Länge der&nbsp;</span> [Hypotenuse](https://de.wikipedia.org/wiki/Hypotenuse) <span style="color:rgb(32, 33, 34);font-family:sans-serif;background-color:rgb(255, 255, 255);">&nbsp;(Seite gegenüber dem rechten Winkel).</span>

<span style="color:rgb(32, 33, 34);font-family:sans-serif;background-color:rgb(255, 255, 255);">Sinus eines Winkels = (Gegenkathete des Winkels) / (Hypothenuse)</span>

```sql
select Sin(0.5*Pi()) as 'SIN' -- 1
Select Cos(PI()) as 'COS' -- 0

select Sin(45) as 'SIN'
select cos(30) as 'COS'
```

<span style="color:rgb(32, 33, 34);font-family:sans-serif;background-color:rgb(255, 255, 255);">Der&nbsp;</span> **Arkussinus** <span style="color:rgb(32, 33, 34);font-family:sans-serif;background-color:rgb(255, 255, 255);">&nbsp;–&nbsp;geschrieben archsin oder asin</span>  <span style="color:rgb(32, 33, 34);font-family:sans-serif;background-color:rgb(255, 255, 255);">&nbsp;– und der&nbsp;</span> **Arkuskosinus** <span style="color:rgb(32, 33, 34);font-family:sans-serif;background-color:rgb(255, 255, 255);">&nbsp;(oder auch&nbsp;</span> **Arkuscosinus**<span style="color:rgb(32, 33, 34);font-family:sans-serif;background-color:rgb(255, 255, 255);">) –&nbsp;geschrieben archcos oder acos&nbsp;</span> <span style="color:rgb(32, 33, 34);font-family:sans-serif;background-color:rgb(255, 255, 255);">– sind&nbsp;</span> [Umkehrfunktionen](https://de.wikipedia.org/wiki/Umkehrfunktion) <span style="color:rgb(32, 33, 34);font-family:sans-serif;background-color:rgb(255, 255, 255);">&nbsp;der (geeignet)&nbsp;</span> [eingeschränkten](https://de.wikipedia.org/wiki/Einschr%C3%A4nkung) [Sinus- bzw. Kosinusfunktion](https://de.wikipedia.org/wiki/Sinus_und_Kosinus) \[Wikipedia\]



```sql
select asin(-1) 'ASIN'
select acos(1) as 'ACOS'
```

**Tangens** <span style="color: rgb(32, 33, 34); font-family: sans-serif; background-color: rgb(255, 255, 255);">und</span> **Kotangens** <span style="color: rgb(32, 33, 34); font-family: sans-serif; background-color: rgb(255, 255, 255);">&nbsp;sind&nbsp;</span> [trigonometrische Funktionen](https://de.wikipedia.org/wiki/Trigonometrische_Funktion) \[Wikipedia\]

  



```sql
select cot(60) as 'COT'
select tan(0) as 'TAN'
```

```sql
select atan(tan(120)) as 'ATAN'
```
